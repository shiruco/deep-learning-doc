## 1.パーセプトロン

ニューラルネットワークの起源となるアルゴリズム

### 重み
重みとは電流でいうところの「抵抗」  
電流の場合は抵抗が低ければ低いほど大きな電流が流れる  
一方、パーセプトロンの場合は重みが大きければ大きいほど大きな信号が流れる　　  
抵抗も重みも信号の流れにくさ（流れやすさ）をコントロールするという点で同じ　　

### バイアス
重み（w1, w2）は入力信号への重要度をコントロールするパラメータ  
バイアスは発火のしやすさをコントロールするパラメータ  
(バイアスも重みに含まれる)

## 2.ニューラルネットワーク

### 活性化関数
入力信号の総和(バイアスを含む)を出力信号に変換する関数のこと  
パーセプトロンでは活性化関数に「ステップ関数」を使用していた。

### ステップ関数
閾値を境にして出力が切り替わる関数のこと

以下のコードの場合、xが0以上なら1、0以下なら0を出力する

```python
import numpy as np
import matplotlib.pylab as plt

%matplotlib inline

def step_func(x):
    return np.array(x > 0, dtype=np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_func(x)

plt.plot(x,y)
plt.ylim(-0.1, 1.2)
plt.show()
```

### シグモイド関数
ニューラルネットワークにおいて昔から利用されてきた活性化関数

#### *h(x) = 1.0 / (1.0 + math.exp(-x))*

```python
import numpy as np
import matplotlib.pylab as plt

%matplotlib inline

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x,y)
plt.ylim(-0.1, 1.2)
plt.show()
```

### シグモイド関数とステップ関数の共通点

- 非線形関数
- シグモイド関数とステップ関数のグラフを比較すると、シグモイドの方が滑らか
- 出力信号は0から1
- 入力信号が小さくなると出力信号は０に近づく
- 入力信号が大きくなると出力信号は1に近く

#### 活性化関数が線形関数ではダメな理由

ニューラルネットワークにおいて層を深くする意味がなくなる

例：
h(x) = cxを活性化関数とすると  
y(x) = h(h(h(x))) が多層ネットワークとなるが、  
y(x) = c ** 3x となり　1層のニュートラルネットワークになってしまう！

### ReLU関数
最近ではシグモイド関数よりもReLU(Rectified Linear Unit)関数が用いられる  
特徴として、入力が0以上であれば入力信号をそのまま出力する

```python
import numpy as np
import matplotlib.pylab as plt

%matplotlib inline

def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x,y)
plt.ylim(-0.1, 1.2)
plt.show()
```
